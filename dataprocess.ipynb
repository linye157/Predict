```python
#读取./data/new_Industrial_data.xlsx文件，忽略第一行
data1 = pd.read_excel('./data/new_Industrial_data.xlsx', skiprows=1)
data2 = pd.read_excel('./data/col(1).xlsx')
```

```python
#查看data1的表头
print(data1.columns)
```

```python
data1_columns_zh = []
for i in data1.columns:
    #查找data2中COLUMN_NAME_TABLE_NAME列中包含data1的列名的索引
    index = data2[data2['COLUMN_NAME_TABLE_NAME'].str.contains(i)].index
    #查找该索引下data2 REMARKS列的值
    data1_columns_zh.append(data2.loc[index, 'REMARKS'].values)

print(data1_columns_zh)
#将data1_columns_zh添加到data1的第一行
data1.loc[-1] = data1_columns_zh
data1.index = data1.index + 1
data1 = data1.sort_index()
#保存添加新行后的数据到源文件
data1.to_excel('./data/new_Industrial_data.xlsx', index=False)
    
```

```python
#读取IndustrialData.xlsx
import pandas as pd
data2 = pd.read_excel('./data/621data.xlsx')
print(data2.head())
```

```python
#查看索引为316的数据
print(data2.loc[316])
```

```python
#如果某一行的数据中有NaN值，则删除该行
data2 = data2.dropna()
```

```python
#删除倒数三列中出现0或者-1的行
data2 = data2[~((data2.iloc[:, -3:] == 0) | (data2.iloc[:, -3:] == -1)).any(axis=1)]


#删除重复行，只保留一个
data2 = data2.drop_duplicates()
```

```python
#将data2按照8.5：1.5的比例分割成训练集和测试集
from sklearn.model_selection import train_test_split
train_data, test_data = train_test_split(data2, test_size=0.15, random_state=42)
#将训练集按照8：2的比例分割成训练集和验证集
train_data, val_data = train_test_split(train_data, test_size=0.2, random_state=42)
#保存三个文件
train_data.to_excel('./data/train_data.xlsx', index=False)
val_data.to_excel('./data/val_data.xlsx', index=False)
test_data.to_excel('./data/test_data.xlsx', index=False)
```
